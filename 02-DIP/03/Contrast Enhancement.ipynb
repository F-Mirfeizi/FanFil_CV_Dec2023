```python
#import 
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('images/main/c.jpg', 0)
```

```python
#showing image using opencv
def show_img_cv(img_title, img):
    cv2.imshow(img_title, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
#show_img_cv('orig_img', img)
```

```python
#showing image using matplotlib
def show_img_plt(img, c_map ='gray', fig_hight=8, fig_width=6):
    plt.figure(figsize=(fig_hight, fig_width), dpi=100)
    plt.imshow(img, cmap=c_map)
```

```python
show_img_plt(img)
```

```python
#plotting histogram
def display_hist_cv(img, channel=0, histsize=256, range=(0, 256)):
    hist_img = cv2.calcHist([img], [channel], None, [histsize], range)
    plt.figure(figsize=(4, 2), dpi=100)
    plt.plot(hist_img)
    plt.show()
```

```python
display_hist_cv(img)
```

```python
#display hist using matplotlib
def display_hist_plt(img, bins=256, range=(0, 256)):
    plt.figure(figsize=(4, 2), dpi=100)
    plt.hist(img.flat, bins=bins, range=range)
    plt.show()
```

```python
display_hist_plt(img)
```

## Contrast Stretching

```python
min_new = 0
max_new = 255

min_old = np.min(img)
max_old = np.max(img)

cs_img = np.uint8(((img-min_old)/(max_old-min_old))*255)
```

```python
show_img_plt(cs_img)
```

```python
display_hist_plt(cs_img)
```

## Histogram Equalization

```python
hi_eq = cv2.equalizeHist(img)
```

```python
display_hist_plt(hi_eq)
```

```python
show_img_plt(hi_eq)
```

## Gamma

```python
gamma = 0.3
gamma_img = np.uint8(((img/255)**gamma)*255)
```

```python
show_img_plt(gamma_img)
```

```python
display_hist_plt(gamma_img)
```

```python
hi_eq_gamma = cv2.equalizeHist(gamma_img)
```

```python
show_img_plt(hi_eq_gamma)
```
