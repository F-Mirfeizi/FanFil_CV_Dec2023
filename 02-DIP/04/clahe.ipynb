```python
#import 
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#showing image using opencv
def show_img_cv(img_title, img):
    cv2.imshow(img_title, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
#showing image using matplotlib
def show_img_plt(img, c_map ='gray', fig_hight=8, fig_width=6):
    plt.figure(figsize=(fig_hight, fig_width), dpi=100)
    plt.imshow(img, cmap=c_map)
```

```python
#display hist using matplotlib
def display_hist_plt(img, bins=256, range=(0, 256)):
    plt.figure(figsize=(4, 2), dpi=100)
    plt.hist(img.flat, bins=bins, range=range)
    plt.show()
```

```python
low_contrast_img = cv2.imread(r'c.jpg', 0)
show_img_plt(low_contrast_img)
```

```python
display_hist_plt(low_contrast_img)
```

```python
he_img = cv2.equalizeHist(low_contrast_img)
show_img_plt(he_img)
```

```python
display_hist_plt(he_img)
```

## CLAHE

```python
clahe_100 =cv2.createCLAHE(clipLimit=100)
clahe_img = clahe_100.apply(low_contrast_img)
show_img_plt(clahe_img)
```

```python
display_hist_plt(clahe_img)
```
